```python
def count_level(filename, level):
    count = 0
    with open(filename, "r") as file:
        for line in file:
            if line.startswith(level):
                count = count + 1
    return count


error_count = count_level("ERROR_1.txt", "ERROR")
warning_count = count_level("ERROR_1.txt", "WARNING")
info_count = count_level("ERROR_1.txt", "INFO")

print("ERROR:", error_count)
print("WARNING:", warning_count)
print("INFO:", info_count)
```

```python
def average_temp(data):
    total = 0
    for t in data:
        total = total + t
    return total / len(data)

with open("30.txt", "r") as file:
    lines = file.readlines()

temps = []
for line in lines:
    temps.append(float(line.strip()))

avg = average_temp(temps)

print("Average Temperature:", avg)

if avg > 25:
    print("High")
elif avg >= 15 and avg <= 25:
    print("Moderate")
else:
    print("Low")
```

```python
def filter_words(ERROR_1_file, ERROR_2_file, min_length):
    with open(ERROR_1_file, "r") as infile, open(ERROR_2_file, "w") as outfile:
        for line in infile:
            words = line.split()
            result = ""

            for word in words:
                if len(word) >= min_length:
                    result = result + word + " "

            outfile.write(result.strip() + "\n")

filter_words("ERROR_1.txt", "ERROR_2.txt", 4)
```

```python
with open("sakthi-1.txt", "w") as file:
    while True:
        value = input("Enter a number (or type 'stop'): ")

        if value == "stop":
            break

        file.write(value + "\n")

total = 0
with open("sakthi-1.txt", "r") as file:
    for line in file:
        total = total + int(line)

print("Total sum:", total)
```

```python
def count_long_lines(filename,limit):
    count = 0
    with open(filename, 'r') as file:
            lines = file.readlines()
    for line in lines:
          if len(line) > limit:
                count += 1
                return count

filename = input("Enter the filename: ")
limit = int(input("Enter the length limit: "))
result = count_long_lines(filename,limit)
print("number of lines longer than", limit, ":", result)
```
