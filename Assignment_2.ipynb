```python
# Create a dictionary for one student with name, age, and a nested dictionary of three subjects and marks
# Example:
   # {
#        "student_id": {
 #           "name": "...",
  #          "age": ...,
   #         "courses": {
    #            "Math": 85,
     #           "Science": 90,
      #          "English": 88
       #     }
        #}
# Print the studentâ€™s name and age, then calculate and print the average mark

student = {
    "S0107": {
        "name": "sakthi",
        "age": 28,
        "courses": {
            "Math": 85,
            "Science": 90,
            "English": 88
        }
    }
}

student_id = "S0107"
S = student [student_id]

print(f"Name: {S['name']}")
print(f"Age: {S['age']}")

marks = S['courses'].values()
average = sum(marks) / len(marks)


print(f"Average Mark: {average:.2f}")
```

```python
department = {
    "Engineering": {
        "budget": 500000,
        "employees": [
            {"name": "sakthi", "salary": 90000},
            {"name": "Abi",   "salary": 85000}
        ]
    }
}


E = department["Engineering"]

total_salary = E["employees"][0]["salary"] + E["employees"][1]["salary"]

print("Total Salary:", total_salary)

if total_salary <= E ["budget"]:
    print("The total salary is within the budget.")
else:
    print("The total salary exceeds the budget.")
```

```python
coordinates = [(10, 20), (5, 15), (7, 3)]

print("First coordinate:", coordinates[0])
print("Last coordinate:", coordinates[-1])

swapped_coordinates = []

for x, y in coordinates:
    swapped_coordinates.append((y, x))


print("Swapped coordinates:", swapped_coordinates)
```

```python
L2 = {"a@gmail.com", "b@gmail.com", "c@gmail.com"}
L3 = {"b@gmail.com", "c@gmail.com", "d@gmail.com"}


common_emails = L2.intersection(L3)
print("Common emails:", common_emails)


all_unique_emails = L2.union(L3)
print("All unique emails:", all_unique_emails)


L2.remove("a@gmail.com")
L3.remove("d@gmail.com")


print("Updated L2:", L2)
print("Updated L3:", L3)
```

```python
library = {
    "book1": {"title": "BIG BOY", "author": "Sakthi", "available": True},
    "book2": {"title": "FUNNY CONCEPT", "author": "ABI", "available": False},
    "book3": {"title": "CRICKET", "author": "VEL", "available": True}
}


borrowing_history = [
    ("book1", "Sakthi"),
    ("book2", "Abi")
]


print("Available books:")
for book_id, details in library.items():
    if details["available"]:
        print("-", details["title"])


print("\nUsers who borrowed books:")
for entry in borrowing_history:
    user_name = entry[1]

    print("-", user_name)
```
